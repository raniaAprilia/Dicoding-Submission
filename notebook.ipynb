## Visualization & Explanatory Analysis

### Pertanyaan 1: Seberapa pengaruh musim pada peminjaman sepeda harian?

```python
day_df.info()
```

```python
plt.figure(figsize=(12, 9))
season_cnt = day_df.groupby(by=["season", "weekday"]).agg({
    "cnt":"sum"
}).reset_index()

sns.barplot(data=season_cnt, x="season", y="cnt", hue="weekday", palette="rocket")
plt.ylabel("Jumlah")
plt.title ("Jumlah total sepeda yang digunakan berdasarkan musim")
plt.legend(title="Weekday", loc="upper right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 9))
season_cnt = hour_df.groupby(by=["season", "weekday"]).agg({
    "cnt":"sum"
}).reset_index()

sns.barplot(data=season_cnt, x="season", y="cnt", hue="weekday", palette="rocket")
plt.ylabel("Jumlah")
plt.title ("Jumlah total sepeda yang digunakan berdasarkan musim")
plt.legend(title="Weekday", loc="upper right")
plt.tight_layout()
plt.show()
```

### Pertanyaan 2: Seberapa besar pengaruh kecepan angin pada peminjaman sepeda pada musim gugur?

```python
working_counts = day_df[day_df['season'] == 'Fall'].groupby(by=['workingday', 'yr']).agg({
    'windspeed': 'sum',
    'cnt': 'sum'
}).reset_index()

sns.barplot(data=working_counts, x='workingday', y='windspeed',hue='yr', palette='viridis')
plt.ylabel('Total Kecepatan Angin')
plt.title('Total Kecepatan Angin berdasarkan Hari Kerja pada Musim Gugur')
plt.tight_layout()
plt.show()
```

```python
working_counts = hour_df[hour_df['season'] == 'Fall'].groupby(by=['workingday', 'yr']).agg({
    'windspeed': 'sum',
    'cnt': 'sum'
}).reset_index()

sns.barplot(data=working_counts, x='workingday', y='windspeed',hue='yr', palette='viridis')
plt.ylabel('Total Kecepatan Angin')
plt.title('Total Kecepatan Angin berdasarkan Hari Kerja pada Musim Gugur')
plt.tight_layout()
plt.show()
```

Penyewaan terbanyak untuk musim Fall terdapat pada hari Kamis, musim Spring pada hari Jumat, musim Winter dan Summer pada hari Sabtu
